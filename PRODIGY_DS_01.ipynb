```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv', delimiter=',', encoding='utf-8', skiprows=4)
df.head()
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
year_columns = [col for col in df.columns if col.isdigit()]
```

```python
country_name = "India"
country_data = df[df['Country Name'] == country_name]

if not country_data.empty:
    population = country_data[year_columns].values.flatten()

    plt.figure(figsize=(10, 5))
    plt.bar(year_columns, population, color='skyblue')
    plt.title(f"{country_name} Population Over the Years")
    plt.xlabel("Year")
    plt.ylabel("Population")
    plt.xticks(ticks=range(0, len(year_columns), 5), labels=[year_columns[i] for i in range(0, len(year_columns), 5)], rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.show()
```
